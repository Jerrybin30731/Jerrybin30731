```python
import openai
import nltk
import time
from nltk.tokenize import sent_tokenize

openai.api_key = (User_API) # put in your API key

with open('cs50csection 1.txt', 'r') as file: # with open('file name.txt', 'r') as file:
    text = file.read()

sentences = sent_tokenize(text)

for sentence in sentences:
    completion = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "你是一個文件翻譯的助手，請將以下的英文技術文件翻譯成繁體中文。"},  #Translate English into mandarin
            {"role": "user", "content": sentence},
        ]
    )
    time.sleep(20) #Set it to every 20 seconds because of the gpt limit
    print(completion.choices[0].message['content'])
```
